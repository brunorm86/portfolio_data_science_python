## Linhas de restrição e anotações

```python
fig = plt.figure(figsize = (16,9))
eixo = fig.add_axes([0,0,1,1])

eixo.plot(df['data'], df['temperatura'], color = 'g')
eixo.set_title('Temperatura no momento', fontsize=25, pad=20) #pad é distancia entre titulo e area do gráfico
eixo.set_ylabel('Temperatura', fontsize=20)
eixo.set_xlabel('Data', fontsize=20)
eixo.legend(['temperatura'], loc = 'lower right', fontsize=15)
eixo.grid(True)

#acrescenta duas linhas tracejadas no gráfico, indicanto os pontos maximo e minimo
eixo.axhline(max(df['temperatura']), color = 'k', linestyle = '--')
eixo.axhline(min(df['temperatura']), color = 'k', linestyle = '--')
```

```python
fig = plt.figure(figsize = (16,9))
eixo = fig.add_axes([0,0,1,1])

eixo.plot(df['data'], df['temperatura'], color = 'g')
eixo.set_title('Temperatura no momento', fontsize=25, pad=20) #pad é distancia entre titulo e area do gráfico
eixo.set_ylabel('Temperatura', fontsize=20)
eixo.set_xlabel('Data', fontsize=20)
eixo.legend(['temperatura'], loc = 'lower right', fontsize=15)
eixo.grid(True)

x1 = df['data'][df['temperatura'].idxmax()]
y1 = max(df['temperatura'])

#Anota os pontos máximo
eixo.annotate("Máximo", xy=(x1,y1), fontsize=20)

#acrescenta duas linhas tracejadas no gráfico, indicanto os pontos maximo e minimo
eixo.axhline(max(df['temperatura']), color = 'k', linestyle = '--')
eixo.axhline(min(df['temperatura']), color = 'k', linestyle = '--')
```

### Criando uma seta no gráfico

```python


fig = plt.figure(figsize = (16,9))
eixo = fig.add_axes([0,0,1,1])

eixo.plot(df['data'], df['temperatura'], color = 'g')
eixo.set_title('Temperatura no momento', fontsize=25, pad=20)
eixo.set_ylabel('Temperatura', fontsize=20)
eixo.set_xlabel('Data', fontsize=20)
eixo.legend(['temperatura'], loc = 'lower right', fontsize=15)
eixo.grid(True)

x1 = df['data'][df['temperatura'].idxmax()]
y1 = max(df['temperatura'])

x2 = df['data'][df['temperatura'].idxmax()-7000] # local de origem da seta
y2 = max(df['temperatura']) - 5

#Anota os pontos máximo
eixo.annotate("Máximo", xy=(x1,y1), fontsize=20, xytext=(x2,y2), arrowprops = dict(facecolor='k'))

x1 = df['data'][df['temperatura'].idxmin()]
y1 = min(df['temperatura'])

x2 = df['data'][df['temperatura'].idxmin()-7000] # local de origem da seta
y2 = min(df['temperatura']) + 5

#Anota os pontos MÍNIMO
eixo.annotate("Mínimo", xy=(x1,y1), fontsize=20, xytext=(x2,y2), arrowprops = dict(facecolor='k'))

#acrescenta duas linhas tracejadas no gráfico, indicanto os pontos maximo e minimo
eixo.axhline(max(df['temperatura']), color = 'k', linestyle = '--')
eixo.axhline(min(df['temperatura']), color = 'k', linestyle = '--')
```

```python
fig = plt.figure(figsize=(8,5))
eixo = fig.add_axes([0,0,1,1])

indice = range(len(temperatura_por_dia_da_semana))

eixo.bar(indice, temperatura_por_dia_da_semana)
eixo.set_title("Temperatura média por dia da semana", fontsize=15, pad=10)
eixo.set_xlabel('Dia da semana', fontsize=15)
eixo.set_ylabel('Temperatura Média', fontsize=15)
eixo.set_xticks(indice)
eixo.set_xticklabels(nome_dias)
```

```python
# adicionando cores às barras

fig = plt.figure(figsize=(8,5))
eixo = fig.add_axes([0,0,1,1])

indice = range(len(temperatura_por_dia_da_semana))

cores = ['black', 'r', 'g', 'b', 'yellow', 'orange', 'magenta'] # lista de cores que será enviada como parâmetro

eixo.bar(indice, temperatura_por_dia_da_semana, color = cores, edgecolor = 'black')
eixo.set_title("Temperatura média por dia da semana", fontsize=15, pad=10)
eixo.set_xlabel('Dia da semana', fontsize=15)
eixo.set_ylabel('Temperatura Média', fontsize=15)
eixo.set_xticks(indice)
eixo.set_xticklabels(nome_dias)
```

```python
# grafico de pizza

fig = plt.figure(figsize=(5,4))
eixo = fig.add_axes([0,0,1,1])

eixo.pie(temperatura_por_dia_da_semana, labels = temperatura_por_dia_da_semana.index)
eixo.set_title('Temperatura por dia da semana', fontsize=15, pad=10)
```

```python
# colocando percentil dentro de cada "fatia'

# grafico de pizza

fig = plt.figure(figsize=(5,4))
eixo = fig.add_axes([0,0,1,1])

eixo.pie(temperatura_por_dia_da_semana, labels = temperatura_por_dia_da_semana.index, autopct='%.2f%%')
eixo.set_title('Temperatura por dia da semana', fontsize=15, pad=10)
```

```python
# "explodir" os finais de semana e pôr sombra

# grafico de pizza

fig = plt.figure(figsize=(5,4))
eixo = fig.add_axes([0,0,1,1])

explodir = [.2,0,0,0,0,.2,.2]

eixo.pie(temperatura_por_dia_da_semana, labels = temperatura_por_dia_da_semana.index, autopct='%.2f%%', explode = explodir, shadow = True)
eixo.set_title('Temperatura por dia da semana', fontsize=15, pad=10)
```

```python
# criando os gráficos

fig = plt.figure(figsize=(16,9))
eixo = fig.add_axes([0,0,1,1])

eixo.scatter(df['comprimento_sépala'], df['largura_sépala']) #o scatter recebe como parametros o eixo X e depois o Y
eixo.set_title('Gráfico de dispersão', fontsize=25, pad=15)
eixo.set_xlabel('Comprimento da sépala', fontsize=15)
eixo.set_ylabel('Largura da sépala', fontsize=15)
```

```python
fig = plt.figure(figsize=(16,9))
eixo = fig.add_axes([0,0,1,1])

eixo.scatter(df['comprimento_sépala'], df['largura_sépala']) #o scatter recebe como parametros o eixo X e depois o Y
eixo.set_title('Gráfico de dispersão', fontsize=25, pad=15)
eixo.set_xlabel('Comprimento da sépala', fontsize=15)
eixo.set_ylabel('Largura da sépala', fontsize=15)

#aumenta o tamanho dos ticks
eixo.tick_params(labelsize=15)
```

```python
# usar cores em cada espécie diferente

fig = plt.figure(figsize=(16,9))
eixo = fig.add_axes([0,0,1,1])

# algoritmo pra atribuir cor e marcador diferente para cada espécie
cores={'Iris-setosa':'r', 'Iris-versicolor':'b', 'Iris-virginica': 'g'}
marcadores = {'Iris-setosa':'x', 'Iris-versicolor':'o', 'Iris-virginica': 'v'}

for especie in df['espécie'].unique():
    tmp = df[df['espécie']==especie]
    eixo.scatter(tmp['comprimento_sépala'], tmp['largura_sépala'], color=cores[especie], marker=marcadores[especie], s=100)
    
#

eixo.set_title('Gráfico de dispersão', fontsize=25, pad=15)
eixo.set_xlabel('Comprimento da sépala', fontsize=15)
eixo.set_ylabel('Largura da sépala', fontsize=15)

#aumenta o tamanho dos ticks
eixo.tick_params(labelsize=15)

#cria legenda para cada cor
eixo.legend(cores, fontsize=20)
```

# Aula 5 - Customizando Histogramas

## Adicionando anotações e linhas horizontais

```python
fig = plt.figure(figsize=(10,8))
eixo = fig.add_axes([0,0,1,1])

mu, sigma = df['comprimento_pétala'].mean(), df['comprimento_pétala'].std()

eixo.hist(df['comprimento_pétala'], bins=20)
eixo.set_title('Histograma', fontsize=15, pad=10)
eixo.set_xlabel('Comprimento da pétala', fontsize=15)
eixo.grid(True)

#Usaremos a função eixo.annotate() para mostrarmos os novos dados no gráfico. Como estamos trabalhando com uma fórmula do Matplotlib nessa função,
#passaremos nosso código entre cifrões ($$), a começar pela média (mu) que receberá a formatação {0:.2f} definindo que queremos exibir duas casas decimais.
#Usaremos \n para pularmos uma linha e, entre um novo par de cifrões, passaremos sigma = {1:.2f}, assim estabelecendo que esse será o próximo valor exibido e que também terá duas casas decimais.

eixo.annotate('$mu = {0:.2f}$\n$sigma = {1:.2f}$'.format(mu, sigma),
               xy=(4.5, 20), fontsize=20)
```

```python
fig = plt.figure(figsize=(10,8))
eixo = fig.add_axes([0,0,1,1])

mu, sigma = df['comprimento_pétala'].mean(), df['comprimento_pétala'].std()

eixo.hist(df['comprimento_pétala'], bins=20)
eixo.set_title('Histograma', fontsize=15, pad=10)
eixo.set_xlabel('Comprimento da pétala', fontsize=15)
eixo.grid(True)

eixo.annotate('$mu = {0:.2f}$\n$sigma = {1:.2f}$'.format(mu, sigma),
               xy=(4.5, 20), fontsize=20)

#inclui uma linha indicando a média
eixo.axvline(mu, color= 'k', linestyle = '--')
eixo.annotate('média', xy=(mu-.7, 28), fontsize=20)
```

```python
fig = plt.figure(figsize=(10,8))
eixo = fig.add_axes([0,0,1,1])

mu, sigma = df['comprimento_pétala'].mean(), df['comprimento_pétala'].std()

eixo.hist(df['comprimento_pétala'], bins=20)
eixo.set_title('Histograma', fontsize=15, pad=10)
eixo.set_xlabel('Comprimento da pétala', fontsize=15)
eixo.grid(True)

eixo.annotate('$\mu = {0:.2f}$\n$\sigma = {1:.2f}$'.format(mu, sigma), # com a '\' antes do mu e do sigma, imprime o simbolo grego, não a palavra
               xy=(4.5, 20), fontsize=20)


eixo.axvline(mu, color= 'k', linestyle = '--')
eixo.annotate('média', xy=(mu-.7, 28), fontsize=20)

eixo.axvline(df['comprimento_pétala'].median(), color= 'r', linestyle = '--')
eixo.annotate('mediana', xy=(df['comprimento_pétala'].median()+.1, 31), fontsize=20)
```
