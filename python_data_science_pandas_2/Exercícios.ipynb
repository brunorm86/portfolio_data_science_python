## Exercícios contagem moedas

```python
import pandas as pd
```

```python
m1 = 'CCcCCccCCCccCcCccCcCcCCCcCCcccCCcCcCcCcccCCcCcccCc'
m2 = 'CCCCCccCccCcCCCCccCccccCccCccCCcCccCcCcCCcCccCccCc'
m3 = 'CccCCccCcCCCCCCCCCCcccCccCCCCCCccCCCcccCCCcCCcccCC'
m4 = 'cCCccCCccCCccCCccccCcCcCcCcCcCcCCCCccccCCCcCCcCCCC'
m5 = 'CCCcCcCcCcCCCcCCcCcCCccCcCCcccCccCCcCcCcCcCcccccCc'
```

### transforma as stringsa em um dicionário
#### o "list" converte a string em uma lista onde cada um dos caracteres vira um item em separado

```python
eventos = {'m1': list(m1), 'm2': list(m2), 'm3': list(m3), 'm4': list(m4), 'm5': list(m5)}
```

```python
eventos
```

#### necessário mandar um dicuinário de "chave: lista de valores" para montar o DF

```python
moedas = pd.DataFrame(eventos) 
moedas
```

### Testes para verificar se a contagem está correta

```python
m1.count('c')
```

#### Criação de uma novo DF que será "acoplado" ao original

```python
#Pela quantidade de index e columns, tems-e adicionar dois items como "data"
df = pd.DataFrame(data = ['Cara','Coroa'], 
                    index = ['c', 'C'], 
                    columns = ['Faces'])
```

```python
df
```

### Agora fazemos a concatenacao deste DF anterior com o o que exibe cada jogada, o "moedas". como dados, botaremos a contagem (.value_counts()) de cada tipo de dado (item) que existe em cada coluna (axis=1). O value_counts() faz a contagem de cada valor diferente que ha no DF. como soh existem dois tipos de dados ('c' e 'C'), ele traz apenas duas linhas para cada coluna.

```python
for item in moedas:
    df = pd.concat([df, moedas[item].value_counts()], axis = 1)
df
```

```python
df.shape
```

## 05 Selecionando as estatísticas descritivas

```python
import pandas as pd
precos = pd.DataFrame([['Feira', 'Cebola', 2.5], 
                        ['Mercado', 'Cebola', 1.99], 
                        ['Supermercado', 'Cebola', 1.69], 
                        ['Feira', 'Tomate', 4], 
                        ['Mercado', 'Tomate', 3.29], 
                        ['Supermercado', 'Tomate', 2.99], 
                        ['Feira', 'Batata', 4.2], 
                        ['Mercado', 'Batata', 3.99], 
                        ['Supermercado', 'Batata', 3.69]], 
                        columns = ['Local', 'Produto', 'Preço'])
precos
```

```python
produtos = precos.groupby('Produto')
produtos.describe().round(2)
```

```python
#ERRADO! AGREGA TODAS AS VARIÁVEIS
produtos.aggregate(['min', 'max', 'sum'])
```

```python
#Agrega apenas pela variável preço
produtos['Preço'].aggregate(['mean', 'std', 'min', 'max']).round(2)
```

```python
#novo DF, "preco_agregado" para guardar a "agregação"
preco_agregado = produtos['Preço'].aggregate(['mean', 'std', 'min', 'max']).round(2)
```

```python
#agora, basta renomear os nomes das colunas
preco_agregado.rename(columns={'mean': 'Média', 'std': 'Desvio-Padrão', 'min': 'Mínimo', 'max': 'Máximo'}, inplace = True)
```

```python
preco_agregado
```
